```python
import numpy as np 
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
## Reading train data
df_train=pd.read_csv('Classification_train.csv')
arr_train=df_train.to_numpy()
print(f"data_train.shape: {arr_train.shape}")
df_train.describe()
```

```python
## Reading test data
df_test=pd.read_csv('classification_test.csv')
arr_test=df_test.to_numpy()
print(f"data_train.shape: {arr_test.shape}")
df_test.describe()
```

```python
## Forming Input and Output labels of training set
x_train=arr_train[:,1:]
print(x_train.shape)
print(x_train)

y_train=arr_train[:,0]
print(y_train.shape)
print(y_train)
```

```python
## Forming Input and Output labels of test set
x_test=arr_test[:,1:]
print(x_test.shape)
print(x_test)

y_test=arr_test[:,0]
print(y_test.shape)
print(y_test)
```

```python
def KNN(x,y,test_ex,k):
    m=x.shape[0]  
    n=x.shape[1]  
    dist=np.zeros(m)
    dist=np.sum((x-np.tile(test_ex,(m,1)))**2,axis=1)
    idx=np.argsort(dist)
    required_idx=idx[0:k]
    nearest_k=np.zeros(k)
    for i in range(k):
        nearest_k[i]=y[required_idx[i]]
    values,c=np.unique(nearest_k,return_counts=True)
    p=np.argmax(c)
    pred=values[p]
    return pred    
```

```python
def accuracy(y_pred,y):
    return((np.sum(y_pred==y))/y.shape)*100     
```

```python
x_cv=x_train[20000:30000,:]
print(x_cv)
y_cv=y_train[20000:30000]
print(y_cv)
```

```python
y_predicted=np.zeros(10000)
for i in range(10000):
    y_predicted[i]=KNN(x_train,y_train,x_cv[i],3)
print(y_predicted) 
```

```python
accuracy(y_predicted,y_cv)
```

```python
y_pred_test=np.zeros(10000)
for i in range(10000):
    y_pred_test[i]=KNN(x_train,y_train,x_test[i],3)
print(y_pred_test) 
csv_file_path = 'KNN_Predicted.csv'
np.savetxt(csv_file_path, y_pred_test, delimiter=',')
print(f"Array has been saved to {csv_file_path}")
```
